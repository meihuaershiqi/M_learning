# 1. 从fundemantals.csv开始！

fundemantals.csv 是这些股票的年报数据

请用数据回答以下问题：

1. S&P500股票在2015年`net income`的均值是多少？最大值比最小值多多少？（每问10分，共计20分）
2. S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？固定资产占总资产比例最小的股票是的代码（ticker symbol）是什么？（每问10分，共计20分）


```python
fund_df = pd.read_csv('fundamentals.csv')
fund_df['Period Ending'] = pd.to_datetime(fund_df['Period Ending'])
fund_df = fund_df.set_index('Period Ending')
df_2015 = fund_df['2015']
df_2016 = fund_df['2016']
```

```python
print('2015平均值：')
df_2015['Net Income'].mean()
```

```python
print('2015最大值比最小值多：')
df_2015['Net Income'].max() - df_2015['Net Income'].min()
```

```python
print('2016比例的均值：')
bili = df_2016['Fixed Assets']/df_2016['Total Assets']
bili.mean()
```

```python
print('2016比例最小的股票是的代码：')
df_2016[df_2016['Fixed Assets']/df_2016['Total Assets']==bili.min()][['Ticker Symbol']]
```

# 2. 加入securities.csv~

securities.csv包含了这些股票的基本信息

1. 请列举出各个sector中的加入时间最早的股票名称（10分）
2. 请列举出每一个州中加入时间最晚的股票名称（10分）


```python
secu_df = pd.read_csv('securities.csv')

# 去除‘Date first added’列的NaN
secu_df = secu_df[pd.notnull(secu_df['Date first added'])]  
```

```python
print('各个sector中的加入时间最早的股票名称:\n')
grouped = secu_df.groupby(secu_df['GICS Sector'])
date_min = grouped['Date first added'].min()
for i in range(len(date_min)):
    data_sym = secu_df[(secu_df['GICS Sector']==date_min.index[i]) & (secu_df['Date first added']==date_min[i])]['Ticker symbol']
    print(f'GICS Sector : {date_min.index[i]}')
    print(f'Date first added(min) : {date_min[i]}')
    
    for j in range(len(data_sym)):
        index = data_sym.index[j]
        print(f'Ticker symbol : {data_sym[index]}')
    print('\n')
```

```python
print('每一个州中加入时间最晚的股票名称:\n')

# 添加州名列
s = secu_df['Address of Headquarters']
s = s.map(lambda x: x[x.rindex(',')+2:])
secu_df['State'] = list(s)

grouped = secu_df.groupby(secu_df['State'])                  #   按州名分组
date_max = grouped['Date first added'].max()
for i in range(len(date_max)):
    data_sym = secu_df[(secu_df['State']==date_max.index[i]) & (secu_df['Date first added']==date_max[i])]['Ticker symbol']
    print(f'State : {date_max.index[i]}')
    print(f'Date first added(max) : {date_max[i]}')
    
    for j in range(len(data_sym)):
        index = data_sym.index[j]
        print(f'Ticker symbol : {data_sym[index]}')
    print('\n')
```

# 3. merge!

现在你需要同时处理来自两个表中的信息了

1. 请思考，合并两个表的信息的时候，我们应该用什么样的准则对其它们（10分）
2. 请列举每个sector在2013-2016年累计Research&Development的总投入（10分）
3. 请列举出每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值（20分）

```python
fund_df = pd.read_csv('fundamentals.csv')
secu_df = pd.read_csv('securities.csv')
secu_df.rename(columns={'Ticker symbol':'Ticker Symbol'},inplace=True)    #  修改列名  'Ticker symbol':'Ticker Symbol'
df = pd.merge(fund_df, secu_df, how='left')                                    #  合并

df['Period Ending'] = pd.to_datetime(df['Period Ending']) #将数据类型转换为日期类型
df = df.set_index('Period Ending') # 将date设置为index
df_1316 = df['2013':'2016']
```

```python
grouped = df_1316.groupby(df_1316['GICS Sector'])
data_sum = grouped['Research and Development'].sum()
sector_list = []
print('每个sector在2013-2016年累计Research&Development的总投入：\n')
for i in range(len(data_sum)):
    sector_list.append(data_sum.index[i])
    print(f'GICS Sector : {data_sum.index[i]}')
    print(f'Research and Development : {data_sum[i]}\n')
```

```python
for sector in sector_list:
    df = df_1316[df_1316['GICS Sector']==sector]                # 单个Sector 的df
    grouped = df.groupby('Security')
    rd_sum = []                                                 # 存放  'R&D' .sum() 和 公司名一一对应的列表
    data_sum = grouped['Research and Development'].sum()        # 每一个Sector中按  'R&D'  分组
    for i in range(len(data_sum)):
        rd_sum.append((data_sum[i],data_sum.index[i]))
    rd_sum = sorted(rd_sum)[-3:]                                # 排序后取最后3个
    print(f'\nGICS Sector : {sector}\n')
    for i in range(1,4):
        print(f'TOP {i} \n Security : {rd_sum[-i][1]} \n Research and Development : {rd_sum[-i][0]} ')
```
